### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.

product_sold_df = combined_sales_data_df.groupby(['region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the sum to "Total_Products_Sold".
product_sold_df.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
product_sold_df.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
products_sold_pivot = combined_sales_data_df.pivot_table(index=['region', 'state', 'city'],
                                                         values='units_sold',
                                                         aggfunc='sum').reset_index()


# Rename the "units_sold" column to "Total_Products_Sold"
products_sold_pivot.rename(columns={'units_sold': 'Total_Products_Sold'}, inplace=True)

# Show the top 5 results.
products_sold_pivot.head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_table_sales = combined_sales_data_df.pivot_table(index='invoice_date',
                                                       values='total_sales',
                                                       aggfunc='sum').reset_index()

# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_sales.rename(columns={'total_sales': 'Total Sales'}, inplace=True)

# Show the table.
pivot_table_sales
```

```python
print(pivot_table_sales.dtypes)
```

```python
print(pivot_table_sales.head())
```

```python
pivot_table_sales['invoice_date'] = pd.to_datetime(pivot_table_sales['invoice_date'])
pivot_table_sales.set_index('invoice_date', inplace=True)
```

```python
print(pivot_table_sales.dtypes)
```

```python
print(pivot_table_sales.index)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.

pivot_table_sales.index = pd.to_datetime(pivot_table_sales.index)

daily_sales = pivot_table_sales.select_dtypes(include='number').resample('D').sum()

print(daily_sales)

# Sort the resampled pivot table in descending order on "Total Sales".
if 'Total Sales' in daily_sales.columns:
    sorted_daily_sales = daily_sales.sort_values(by='Total Sales', ascending=False)
    print(sorted_daily_sales)
else:
    print("Column 'Total Sales' not found in the DataFrame.")
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.

combined_sales_data_df.set_index('invoice_date', inplace=True)
weekly_sales_resampled_pivot_table = combined_sales_data_df.resample('W').sum()

# Sort the resampled pivot table in descending order on "Total Sales".
weekly_sales_resampled_pivot_table.sort_values(by='total_sales', ascending=False)
weekly_sales_resampled_pivot_table
```
